```python
df = pd.read_csv("employee_data.csv")
df2 = df.copy()
df["tax"] = df2["Salary"].apply(lambda x:"20%"if x>=50000 else "10%")
df
```

```python
#assign
df2.assign(new_income = df2["Salary"]*1.1)
```

```python
#replace
df2["Department"] =df2["Department"].replace("H R","HeadR")
df2
```

```python
df2.rename(columns={"Salary":"Income"})
```

```python
df2 = df.copy()
new_col_order = list(df.columns)
new_col_order.remove('ID')
new_col_order.append('ID')
new_col_order
df2[new_col_order]
```

```python
df2 = df.copy()
df2 = df2.drop_duplicates() 
df2.fillna(0)
df2=df2.sort_values("Salary")
df2.reset_index(drop = True)
df2.to_csv("Sorted_Data")
```

```python
df2.groupby("Department")["Salary"].mean()
df2.groupby("Department")["Salary"].min()
```

```python
df2.groupby("Department")["Salary"].agg(avg_sal="mean",max_sal="max",min_sal="min")
```

```python
df2.groupby("Department").agg({
    "Salary":"mean",
    "Age":"min"
})
```

```python
df2.groupby("Department").agg(
    max_sal = ("Salary":"mean"),
    "Age":"min"
)
```
