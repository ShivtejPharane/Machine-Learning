### Association Rule Minning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
transactions=[
    ['Bread','Chese','Egg','Juice'],
    ['Bread','Chese','Juice'],
    ['Bread','Milk','Youghurt'],
    ['Bread','Milk','Juice'],
    ['Milk','Chese','Juice']
]
```

```python
transactions2=[
    ['E','A','D','B'],
    ['D','A','C','B','E'],
    ['C','A','B','A'],
    ['B','A','D'],
    ['D'],
    ['D','B'],
    ['A','D','E'],
    ['B','C']
]
```

### Preprocess the Data

```python
#Encode the Transactions
from mlxtend.preprocessing import TransactionEncoder
encoder = TransactionEncoder()
encoded_Transaction = encoder.fit(transactions2).transform(transactions2)
#create the data frame from the encoded Transaction
```

```python
df=pd.DataFrame(encoded_Transaction,columns=encoder.columns_)
df
```

### Find the Rules

```python
from mlxtend.frequent_patterns import fpgrowth
fpgrowth(df,use_colnames=True)
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df,use_colnames=True)
```
