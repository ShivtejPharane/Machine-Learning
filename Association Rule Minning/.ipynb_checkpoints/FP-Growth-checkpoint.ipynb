### Association Rule Minning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
transactions=[
    ['Bread','Chese','Egg','Juice'],
    ['Bread','Chese','Juice'],
    ['Bread','Milk','Youghurt'],
    ['Bread','Milk','Juice'],
    ['Milk','Chese','Juice']
]
```

### Preprocess the Data

```python
#Encode the Transactions
from mlxtend.preprocessing import TransactionEncoder
encoder = TransactionEncoder()
encoded_Transaction = encoder.fit(transactions).transform(transactions)
#create the data frame from the encoded Transaction
```

```python
df=pd.DataFrame(encoded_Transaction,columns=encoder.columns_)
df
```

### Find the Rules

```python
from mlxtend.frequent_patterns import apriori
apriori(df,min_support=0.5,use_colnames=True)
```
