```python
df.drop(['Ash','Magnesium','Color_Intensity'],axis=1,inplace=True)
```

### Data Prepration

```python
# Split the data
x = df.drop('Customer_Segment',axis=1)
y = df['Customer_Segment']
```

```python
# change the classes 
# 0->1 1->2 2->3
y=y-1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

### Build the MOdels

```python
from sklearn.ensemble import AdaBoostClassifier
def create_adaboost_model():
    model = AdaBoostClassifier()
    model.fit(x_train,y_train)
    return model
    
```

```python
from sklearn.ensemble import GradientBoostingClassifier
def create_gradient_model():
    model = GradientBoostingClassifier(learning_rate=0.1,max_depth=4,n_estimators=10,verbose=1)
    model.fit(x_train,y_train)
    return model
```

```python
from xgboost import XGBClassifier
def create_XGB_model():
    model=XGBClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
from lightgbm import LGBMClassifier
def create_lgbm_model():
    model = LGBMClassifier(learning_rate=0.1,n_estimators=10)
    model.fit(x_train,y_train)
    return model
```

```python
from catboost import CatBoostClassifier
def create_catboot_model():
    model = CatBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

### Evaluate Model

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
def evaluate_model(model,name,x,y_true):
    y_pred = model.predict(x)
    accuracy = accuracy_score(y_true,y_pred)
    precision = precision_score(y_true,y_pred,average='micro')
    f1 = f1_score(y_true,y_pred,average='micro')
    recall = recall_score(y_true,y_pred,average='micro')
    return name,accuracy,precision,f1,recall
```

```python
#create all models
model_adaboost =  create_adaboost_model()
model_gredient_boost = create_gradient_model()
model_xgb = create_XGB_model()
model_light_gbm = create_lgbm_model()
model_cat_boost = create_catboot_model()
```

```python
#collect all models collecion : 
models=[
    {'name':'AdaBoost','model':model_adaboost},
    {'name':'Gredient','model':model_gredient_boost},
    {'name':'XGBoost','model':model_xgb},
    {'name':'Light_boost','model':model_light_gbm},
    {'name':'Cat Boost','model':model_cat_boost}
]
```

```python
evaluation_metrics=[]
for info in models:
    evaluation_metrics.append(evaluate_model(info['model'],info['name'],x_train,y_train))

pd.DataFrame(evaluation_metrics,columns=['MOdelName','Accuracy','Precision','F1','Recall'])
```

```python
evaluation_metrics=[]
for info in models:
    evaluation_metrics.append(evaluate_model(info['model'],info['name'],x_test,y_test))

pd.DataFrame(evaluation_metrics,columns=['MOdelName','Accuracy','Precision','F1','Recall'])
```
