```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```

```python
# Get statistical info
df.describe()
```

### Data Cleansing

```python
# Since RowNumber CustomerId Surname not useful so we drop them
df.drop(['RowNumber','CustomerId','Surname'],inplace=True,axis=1)
```

```python
#convert Geography to numrical
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
Encoder.fit(df['Geography'])
df['Geography'] = Encoder.transform(df['Geography'])
#convert Gender to numrical
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
Encoder.fit(df['Gender'])
df['Gender'] = Encoder.transform(df['Gender'])
```

```python
df.info()
```

```python
#Check if any Missing values
df.isna().sum()
```

```python
##corr -> corelation coeffient :[-1 to 1]
# if -1 : Perfect negative corelation
# if  1 : perfect +ve corelation
# 0 : no relation (less important)
df.corr()['Exited'].abs().sort_values()
```

```python
#check the dataset is balanced or not 
df['Exited'].value_counts()
```

### Data Preprossecing

```python
# Split the data into x and y
x = df.drop('Exited',axis=1)
y = df['Exited']
```

### Oversampling using SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_new,y_new = smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y_new,train_size=0.8,random_state=42)
```

### Train the Model

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV(max_iter=20000)
model.fit(x_train,y_train)
```

 ### Model Evaluation

```python
#predict the x_test dataset
y_pred = model.predict(x_test)
```

### Confusion metrics

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# Find the Positions 
tn,fp,fn,tp = cm.ravel()
print(f"tn:{tn}, fp:{fp}, fn:{fn}, tp:{tp}")
```

```python
# Accuracy 
from sklearn.metrics import accuracy_score
accuracy = (tp+tn)/(tn+fp+fn+tp)
print(f"Accuracy : {accuracy *100:.2f}")
a = accuracy_score(y_test,y_pred)
print(f"Accuracy : {a*100:.2f}")
```

```python
#Precision
from sklearn.metrics import precision_score
precision = precision_score(y_test,y_pred)
print(f"Precision : {precision:.2f}")
pr = tp/(tp+fp)
print(f"Precision : {precision:.2f}")
```

```python
#Recall
from sklearn.metrics import recall_score
recall = recall_score(y_test,y_pred)
print(f"recall : {recall:.2f}%")
r = tp/(tp+fn)
print(f"recall : {r:.2f}%")
```

```python
#F1 Score
from sklearn.metrics import f1_score
f1_score = f1_score(y_test,y_pred)
print(f"F1 Score = {f1_score:.2f}")
f1 = 2*(precision*recall)/(precision+recall)
print(f"F1 : {f1_score:.2f}")
```

```python
#roc value
from sklearn.metrics import roc_auc_score
roc = roc_auc_score(y_test,y_pred)
print(f"ROC-AUC = {roc:.2f}")
```

```python
#Roc curve
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
plt.scatter(fpr,tpr)
plt.plot(fpr,tpr,color='orange')
```
