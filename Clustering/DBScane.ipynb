### DBSCAN


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Grnrate the Data

```python
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=1000,noise=0.1,random_state=42)
x
```

### Visualize the data

```python
plt.scatter(x[:,0],x[:,1])
```

### 3 Build using the Kmeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
plt.scatter(x[:,0][kmeans.labels_==0],x[:,1][kmeans.labels_==0])
plt.scatter(x[:,0][kmeans.labels_==1],x[:,1][kmeans.labels_==1]) 
plt.scatter( 1.18674998, -0.06472734)
plt.scatter(-0.22070083,  0.57340669)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

### Build the Hierarcgical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
algomerative = AgglomerativeClustering(n_clusters=2)
algomerative.fit(x)
```

```python
plt.scatter(x[:,0][algomerative.labels_==0],x[:,1][algomerative.labels_==0])
plt.scatter(x[:,0][algomerative.labels_==1],x[:,1][algomerative.labels_==1])
```

### Build With the DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.1,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
# Visulaization
plt.scatter(x[:,0][dbscan.labels_==0],x[:,1][dbscan.labels_==0])
plt.scatter(x[:,0][dbscan.labels_==1],x[:,1][dbscan.labels_==1])
plt.scatter(x[:,0][dbscan.labels_==-1],x[:,1][dbscan.labels_==-1])
```

### Evaluate The Cluster

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,dbscan.labels_)
```
