## Heirarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the data 

```python
df = pd.read_csv('Mall_Customers.csv')
```

### EDA

```python
df.info()
```

```python
# remove unwanted columns
df.drop(['CustomerID','Gender','Age'],axis=1,inplace=True)
```

```python
# Rename the columns
df.rename({
    "Annual Income (k$)" :"income",
    "Spending Score (1-100)" :"spending score"
},axis=1,inplace=True)
```

### Build The Clusters Using The Heirarchical Clusters

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5)
clustering.fit(df)
```

```python
## Get the clusters 
clustering.labels_
```

### Cluster visualization

```python
markers=['+','s','*','^','>']
for k in range(5):
    plt.scatter(
        df['income'][clustering.labels_==k],
        df['spending score'][clustering.labels_==k],
        marker=markers[k]
    )

plt.xlabel('Income')
plt.ylabel('Spending score')
plt.title('Aglimorative Clusters')
```

### Dendrogram Visualization

```python
from scipy.cluster.hierarchy import linkage , dendrogram
linkage_single = linkage(df,method='single')
_=dendrogram(linkage_single)
```

```python
linkage_complete =  linkage(df,method='complete')
_= dendrogram(linkage_complete)
```

```python
linkage_average = linkage(df,'average')
_=dendrogram(linkage_average)
```
