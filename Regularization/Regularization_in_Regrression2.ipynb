```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the Dataset
df = pd.read_csv('data_set_for_overfitting.csv')
```

```python
x = df.drop('y',axis=1)
y = df['y']
```

```python


#Spliting the Dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
def create_Linear_regression_model():
    model = LinearRegression()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.linear_model import Lasso
def create_Lasso_model():
    model=Lasso(alpha=5)
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.linear_model import Ridge
def create_Ridge_model():
    model=Ridge(alpha=5)
    model.fit(x_train,y_train)
    return model
    
```

```python
from sklearn.linear_model import ElasticNet
def create_Elastic_Net():
    model=ElasticNet(alpha=5)
    model.fit(x_train,y_train)
    return model
```

```python
linear_regression_model = create_Linear_regression_model()
lasso_regression_model = create_Lasso_model()
ridge_regression_model = create_Ridge_model()
elasticnet_regression_model = create_Elastic_Net()
models=[
    {"model_name":"Linear Model","model":linear_regression_model},
    {"model_name":"Lasso Model","model":lasso_regression_model},
    {"model_name":"Ridge Model","model":ridge_regression_model},
    {"model_name":"Elastic Net Model","model":ridge_regression_model}
]
```

```python
##Evaluation of model
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,root_mean_squared_error,r2_score
def evaluation_of_metrices(model,model_name,x,y):
    y_pred = model.predict(x)
    mae = mean_absolute_error(y,y_pred)
    mse = mean_squared_error(y,y_pred)
    mape = mean_absolute_percentage_error(y,y_pred)
    rmse = root_mean_squared_error(y,y_pred)
    r2 = r2_score(y,y_pred)
    return mae,mse,mape,rmse,r2
```

```python
#For the Training Data
elvaluation_for_training=[]
#evaluate every model
for model_info in models:
    metrice = evaluation_of_metrices(model=model_info["model"],model_name=["model_name"],x=x_train,y=y_train)
    elvaluation_for_training.append(metrice)

pd.DataFrame(elvaluation_for_training,columns=["MAE","MSE","MAPE","RMSE","R2"])
```

```python
#For Testing Data
elvaluation_for_training=[]
#evaluate every model
for model_info in models:
    metrice = evaluation_of_metrices(model=model_info["model"],model_name=["model_name"],x=x_test,y=y_test)
    elvaluation_for_training.append(metrice)

pd.DataFrame(elvaluation_for_training,columns=["MAE","MSE","MAPE","RMSE","R2"])
```
